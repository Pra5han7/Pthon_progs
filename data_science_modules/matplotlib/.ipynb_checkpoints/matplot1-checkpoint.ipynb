```python
import matplotlib.pyplot as plt #importing
```

```python
#to make a plot after running a cell
%matplotlib inline
#use plt.show() if using any other system than jupyter
```

```python
import numpy as np
x= np.linspace(1,5,5)
y= x**2
```

```python
x
```

```python
y
```

```python
#FUNCTIONAL PLOT
plt.plot(x,y)

#if on other systems tha jupyter
#use plt.show() after plots
```

```python
# add labels and title
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('square function')
```

```python
#subplots
plt.subplot(1,2,1) #(1,2,1) (row,col,plot_num)
plt.plot(x,y,'r') #r is for red

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object Oriented Method
#create a figure object
fig=plt.figure()
```

figure object can think of an empty canvas on which you can create axes.

```python
fig
```

```python
axes=fig.add_axes([0.1,0.1,0.8,0.8]) #axes takes a list of left,bottom,width,height
```

```python
fig
```

```python
#three steps combined
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8]) 
axes.plot(x,y)
```

__subplot in OO method__

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3]) #a subplot

axes1.set_title('X vs Y')
axes1.plot(x,y)

axes2.set_title('Y vs X')
axes2.plot(y,x)
```
