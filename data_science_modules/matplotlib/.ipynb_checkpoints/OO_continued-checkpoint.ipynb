```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x=np.linspace(1,5,5)
y=x**2
z=x**3
a=x**1/2
```

```python
print(x)
print(y)
print(z)
print(a)
```

```python
#recap

fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
```

__subplot__

```python
fig,axes=plt.subplots()
```

```python
#checking axes type
axes
```

now subplots takes in arguments for rows and column and can produce plots based off them

```python
fig,axes = plt.subplots(nrows=1,ncols=2) #can use shift tab to see args it takes
```

```python
#checking type of axes
axes
```

axes here is an array [list] means we can iterate through it

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
for current in axes:
    current.plot(x,y)
```

we can also index through axes

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('square')

axes[1].plot(x,z)
axes[1].set_title('cube')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3)
axes[0].plot(x,y)
axes[0].set_title('square')

axes[1].plot(x,z)
axes[1].set_title('cube')

axes[2].plot(x,a)
axes[2].set_title('square_root')
```

to fix the plot over lapping use tight_layout() as below

```python
fig,axes = plt.subplots(nrows=1,ncols=3)
axes[0].plot(x,y)
axes[0].set_title('square')

axes[1].plot(x,z)
axes[1].set_title('cube')

axes[2].plot(x,a)
axes[2].set_title('square_root')
plt.tight_layout()
```

__picture size and DPI__

```python
fig = plt.figure(figsize=(3,2))
axes= fig.add_axes([0,0,1,1])

axes.plot(x,y)
```

```python
fig = plt.figure(figsize=(8,2))
axes= fig.add_axes([0,0,1,1])

axes.plot(x,y)
```

```python
fig = plt.figure(figsize=(3,2),dpi=100)
axes= fig.add_axes([0,0,1,1])

axes.plot(x,y)
```

```python
fig = plt.figure(figsize=(3,2),dpi=500)
axes= fig.add_axes([0,0,1,1])

axes.plot(x,y)
```

see how changing dpi changes the fig properties

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2)) #default dpi
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

__legend()__

legend comes handy when you have multiple plot on a same canvas

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.plot(x,z)
ax.plot(x,a)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Graph')
```

it is difficult to get what plot is what at first sight

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,label='x square')
ax.plot(x,z,label='x cube')
ax.plot(x,a,label='x sqrt')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Graph')

ax.legend(loc=10) # 0 =>loc=<10
```

legend takes in arg loc you can adjust the loc of legend from there loc=0 is best. there are best,upper right,upper left etc choices avalable
