__appearance__

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(1,5,5)
y = x+1
z = 2*x+1
```

```python
print(x,y,z)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x)
ax.plot(y)
ax.plot(z)
```

__color__

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,color='red')
ax.plot(y,color='blue')
ax.plot(z,color='green')
```

__line style and width__

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,color='red',lw=1)
ax.plot(y,color='blue',lw=2)
ax.plot(z,color='green',linewidth=3) #lw can also be written as linewidth
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.set_title('linear lines')
ax.set_xlabel('X')
ax.set_label('F(X)')

ax.plot(x,label='x',color='red',lw=1,ls='-') 
ax.plot(y,label='x+1',color='blue',lw=2,linestyle='--') # -- for dashed line, also linestyle and ls can be used interchangeably
ax.plot(z,label='2x+1',color='green',linewidth=3,ls='-.') # - for solid lines

ax.legend()
```

__markers__

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.set_title('linear lines')
ax.set_xlabel('X')
ax.set_label('F(X)')

ax.plot(x,label='x',color='red',lw=1,ls='-',marker='o') # - for solid lines
ax.plot(y,label='x+1',color='blue',lw=2,linestyle='--',marker='+',markersize=10) # -- for dashed line, also linestyle and ls can be used interchangeably
ax.plot(z,label='2x+1',color='green',linewidth=3,ls='-.',marker='*',markersize=10) 

ax.legend()
```

__plot limits__

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.set_title('linear lines')
ax.set_xlabel('X')
ax.set_label('F(X)')

ax.plot(x,label='x',color='red',lw=1,ls='-',marker='o') # - for solid lines
ax.plot(y,label='x+1',color='blue',lw=2,linestyle='--',marker='+',markersize=10) # -- for dashed line, also linestyle and ls can be used interchangeably
ax.plot(z,label='2x+1',color='green',linewidth=3,ls='-.',marker='*',markersize=10) 
ax.set_xlim([1,4]) #takes a list of lower and upper bound
ax.set_ylim([1,8])
ax.legend()
```
